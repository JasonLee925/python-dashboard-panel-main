## (0) Some minor data preprocessing

```python
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/ df['population'], 0)
```

## Preparing widgets

```python
# Define Panel widgets
t_year_slider = pn.widgets.IntSlider(name='Year slider', start=1750, end=2020, step=5, value=1850)
t_year_slider
```

```python
# Radio buttons for CO2 measures
t_yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['co2', 'co2_per_capita',],
    button_type='danger'
)
t_yaxis_co2
```

```python
continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']
```

## (1) Line Chart: CO2 emission over time by continent

```python
def fn_co2_plot(t_year_slider, df, continents, t_yaxis_co2):
    tdf = (df[
            (df['year'] <= t_year_slider) &
            (df['country'].isin(continents))
        ]
        .groupby(['country', 'year'], as_index=False)[t_yaxis_co2].mean() 
        .sort_values(by='year')
    )

    # print(tdf.head())

    co2_plot = tdf.hvplot.line(
        x = 'year',
        by='country',
        y=t_yaxis_co2, 
        line_width=2, 
        title="CO2 emission by continent",
        sizing_mode="stretch_width",
        legend='right',
    )
    return co2_plot


linechart = pn.WidgetBox(pn.Column(pn.Row(t_yaxis_co2, t_year_slider),
                         pn.Row(pn.bind(fn_co2_plot, t_year_slider, df, continents, t_yaxis_co2)),
                         align="start",
                         sizing_mode="stretch_width"))
# linechart
```

## (2) Table - CO2 emission over time by continent 

```python
def fn_co2_table(t_year_slider, df, continents, t_yaxis_co2):
    tdf = (df[
            (df['year'] <= t_year_slider) &
            (df['country'].isin(continents))
        ]
        .groupby(['country', 'year'], as_index=False)[t_yaxis_co2].mean() 
        .sort_values(by='year')
        .reset_index(drop=True)
    )

    # print(tdf.head())

    co2_table = tdf.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10) 

    return co2_table


table = pn.WidgetBox(pn.Column(pn.Row(t_yaxis_co2, t_year_slider),
                         pn.Row(pn.bind(fn_co2_table, t_year_slider, df, continents, t_yaxis_co2)),
                         align="start",
                         sizing_mode="stretch_width"))
# table
```

## (3) Scatter Plot: CO2 vs GDP 

```python
def fn_co2_vs_gdp_scatterplot(t_year_slider, df, continents):
    tdf = (df[
                (df['year'] == t_year_slider) &
                (~ (df['country'].isin(continents)))
            ]
            .groupby(['country', 'year', 'gdp_per_capita'], as_index=False)['co2'].mean() 
            .sort_values(by='year')
        )

    # print(tdf.head())

    scatter_plot = tdf.hvplot(
        x='gdp_per_capita', 
        y='co2', 
        by='country', 
        size=80, 
        alpha=0.8,
        legend=False, 
        kind='points', 
    ) 

    return scatter_plot

# fn_co2_vs_gdp_scatterplot(t_year_slider, df, continents)

scattorplot = pn.WidgetBox(pn.Column(t_year_slider,
                         pn.Row(pn.bind(fn_co2_vs_gdp_scatterplot, t_year_slider, df, continents,)),
                         align="start",
                         sizing_mode="stretch_width"))
# scattorplot
```

## (4) Bar chart with CO2 sources by continent

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(
    name='Y axis', 
    options=['coal_co2', 'oil_co2', 'gas_co2'], 
    button_type='danger'
)

continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

def fn_co2_source_bar_plot(t_year_slider, df, continents, yaxis_co2_source):
    tdf =(
        df[
            (df['year'] == t_year_slider) &
            (df['country'].isin(continents))
        ]
        .groupby(['year', 'country'], as_index=False)[yaxis_co2_source].sum()
        .sort_values(by='year')  
    )

    bar_chart = tdf.hvplot(
        kind='bar', 
        x='country', 
        y=yaxis_co2_source, 
        title='CO2 source by continent',
    )

    return bar_chart


barchart = pn.WidgetBox(pn.Column(pn.Row(yaxis_co2_source, t_year_slider),
                         pn.Row(pn.bind(fn_co2_source_bar_plot, t_year_slider, df, continents_excl_world, yaxis_co2_source)),
                         align="start",
                         sizing_mode="stretch_width"))
# barchart
```

## Creating Dashboard

```python
#Layout using Template
template = pn.template.FastListTemplate(
    title='World CO2 emission dashboard', 
    sidebar=[pn.pane.Markdown("# CO2 Emissions and Climate Change"), 
             pn.pane.Markdown("#### Carbon dioxide emissions are the primary driver of global climate change. Its widely recognised that to avoid the worst impacts of climate change, the world needs to urgently reduce emissions. But, how this responsibility is shared between regions, countries, and individuals has been an endless point of contention in international discussions."), 
             pn.pane.PNG('climate_day.png', sizing_mode='scale_both'),
             pn.pane.Markdown("## Settings"),   
             t_year_slider],
    main=[pn.Row(pn.Column(t_yaxis_co2, 
                           pn.bind(fn_co2_plot, t_year_slider, df, continents, t_yaxis_co2)), 
                pn.bind(fn_co2_table, t_year_slider, df, continents, t_yaxis_co2)), 
          pn.Row(pn.bind(fn_co2_vs_gdp_scatterplot, t_year_slider, df, continents), 
                 pn.Column(yaxis_co2_source, 
                           pn.bind(fn_co2_source_bar_plot, t_year_slider, df, continents_excl_world, yaxis_co2_source)))],
    accent_base_color="#A01346",
    header_background="#A01346",
    main_max_width="100%",
)
template.show()
# template.servable();
```
